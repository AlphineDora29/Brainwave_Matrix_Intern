###### Visualizing Potential Outliers with Boxplots

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define numerical columns
num_cols = ["age", "quantity", "price_per_unit", "total_amount"]

# Set figure size
plt.figure(figsize=(12, 8))

# Create subplots for each numerical column
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot of {col}")

plt.tight_layout()
plt.show()
```

###### Detecting Outliers Using the IQR Method

```python
# Function to detect outliers using the IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = data[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile Range
    lower_bound = Q1 - 1.5 * IQR  # Lower threshold
    upper_bound = Q3 + 1.5 * IQR  # Upper threshold
    
    # Find outliers
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Apply the function to each numerical column
outliers_dict = {col: detect_outliers_iqr(df, col) for col in num_cols}

# Display the count of outliers in each column
outlier_counts = {col: len(outliers_dict[col]) for col in num_cols}
outlier_counts
```

###### Summarizing the Results

```python
# Print a summary
for col, count in outlier_counts.items():
    print(f"{col}: {count} outliers detected")

if all(count == 0 for count in outlier_counts.values()):
    print("\nConclusion: No outliers detected in this dataset based on the IQR method.")
else:
    print("\nConclusion: Some outliers were detected. Further investigation may be needed.")
```
