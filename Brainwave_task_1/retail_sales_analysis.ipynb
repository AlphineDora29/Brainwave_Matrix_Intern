```python
# Convert 'date' to datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')


# Convert categorical columns to 'category' type
df['gender'] = df['gender'].astype('category')
df['product_category'] = df['product_category'].astype('category')

print(df.dtypes)
```

### Calculate Key Sales Metrics

```python
# Total Revenue
total_revenue = df['total_amount'].sum()
print(f"Total Revenue: ${total_revenue:,.2f}")

# Average Transaction Value
average_transaction = df['total_amount'].mean()
print(f"Average Transaction Value: ${average_transaction:,.2f}")

# Extract Month-Year for trend analysis
df['month_year'] = df['date'].dt.to_period('M')

# Monthly Sales Trends
monthly_sales = df.groupby('month_year')['total_amount'].sum().reset_index()

# Convert period to string for plotting
monthly_sales['month_year'] = monthly_sales['month_year'].astype(str)
print(monthly_sales.head())  # Display first few rows of trend data
```

### Visualize Monthly Sales Trends

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot Monthly Sales Trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='month_year', y='total_amount', marker='o', linestyle='-')

# Improve readability
plt.xticks(rotation=45)
plt.xlabel('Month-Year')
plt.ylabel('Total Sales ($)')
plt.title('Monthly Sales Trends')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Find highest and lowest revenue months
highest_month = monthly_sales.loc[monthly_sales['total_amount'].idxmax()]
lowest_month = monthly_sales.loc[monthly_sales['total_amount'].idxmin()]

print(f"Highest Revenue Month: {highest_month['month_year']} (${highest_month['total_amount']:,})")
print(f"Lowest Revenue Month: {lowest_month['month_year']} (${lowest_month['total_amount']:,})")

# Plot the monthly sales trend
plt.figure(figsize=(10, 5))
plt.plot(monthly_sales["month_year"], monthly_sales["total_amount"], marker="o", linestyle="-", color="b")

# Formatting the plot
plt.title("Monthly Sales Trends", fontsize=14)
plt.xlabel("Month-Year", fontsize=12)
plt.ylabel("Total Sales ($)", fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid for better readability
plt.show()  # Display the plot
```

## Data Consistency Check

```python
print(df.columns)  # Display column names
```

```python
# Check for missing values
print(df.isnull().sum())

# Check for duplicate rows
print(f"Duplicate rows: {df.duplicated().sum()}")

# Verify if 'total_amount' and 'calculated_total' match
df['mismatch'] = df['total_amount'] != df['calculated_total']
print(f"Mismatched values: {df['mismatch'].sum()}")

# Drop the mismatch column if not needed
df.drop(columns=['mismatch'], inplace=True)
```

## Sales Insights

### Top-Selling Product Categories

```python
top_categories = df.groupby('product_category', observed=False)['total_amount'].sum().sort_values(ascending=False)
print(top_categories)
```

```python
gender_sales = df.groupby('gender', observed=False)['total_amount'].sum()
print(gender_sales)
```

### Sales Trends by Age Group

## Data Visualization

### Bar Chart - Total Sales by Product Category

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.barplot(
    x=top_categories.index, 
    y=top_categories.values, 
    hue=top_categories.index,  # Assign hue to match x values
    palette="viridis", 
    legend=False  # Disable legend for clarity
)
plt.xticks(rotation=45)
plt.title("Total Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales ($)")
plt.show()
```

## Customer Behavior Analysis

### Customer Spending Patterns

```python
# Group by customer_id and sum total sales per customer
customer_spending = df.groupby("customer_id")["total_amount"].sum().sort_values(ascending=False)

# Display top 10 customers
customer_spending.head(10)


plt.figure(figsize=(10,5))
customer_spending.head(10).plot(kind='bar', color='royalblue')
plt.title("Top 10 Customers by Total Spending")
plt.xlabel("Customer ID")
plt.ylabel("Total Spending ($)")
plt.show()
```

### Gender-Based Sales Analysis

```python
gender_sales = df.groupby("gender", observed=False)["total_amount"].sum()
plt.figure(figsize=(6,4))

sns.barplot(x=gender_sales.index, y=gender_sales.values, hue=gender_sales.index, palette="coolwarm", legend=False)

plt.title("Total Sales by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Sales ($)")
plt.show()
```

### Age Group Analysis

```python
age_sales = df.groupby("age_group", observed=False)["total_amount"].sum()

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.barplot(x=age_sales.index, y=age_sales.values, hue=age_sales.index, palette="viridis", legend=False)

plt.title("Total Sales by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Sales ($)")
plt.show()
```
